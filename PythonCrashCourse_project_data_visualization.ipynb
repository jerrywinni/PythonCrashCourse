In this project, we'll use `Matplotlib` and `Plotly` two third-party library.

````python
import matplotlib.pyplot as plt

squares = [1, 4, 9, 16, 25]

""" The variable fig represents the entire figure, which is the collection of plots that are generated. 
The variable ax represents a single plot in the figure; 
this is the variable weâ€™ll use most of the time when defining and customizing a single plot."""
fig, ax = plt.subplots() 
ax.plot(squares)

plt.show()
````

*   `import matplotlib.pyplot as plt`: This line is importing the pyplot module from the matplotlib library and giving it an alias plt. This is a convention in the Python community to shorten the module name. pyplot is a collection of functions that help generate plots and figures like in MATLAB.

*   `squares = [1, 4, 9, 16, 25]`: Here, you're creating a Python list named squares which contains the squares of the first five integers.

*   `fig, ax = plt.subplots()`: This line is initializing a figure and a set of subplots. Here, since no arguments are passed to subplots(), it creates a single plot. fig is a Figure object which is the whole window or page that everything is drawn on. ax is an Axes object or a subplot, which is an individual plot within the figure. You can have multiple subplots (Axes) in a single Figure.

*   `ax.plot(squares)`: This line is plotting the squares list on the Axes object ax. Since only y values (squares) are given, x values will be automatically created as a range from 0 to N-1, where N is the number of items in squares (so here, 0 to 4).

*   `plt.show()`: This line is calling the show() function which displays the plot on the screen. Until this function is called, the plot is just prepared in the background but not displayed.

````python
import matplotlib.pyplot as plt

input_values = [1, 2, 3, 4, 5]
squares = [1, 4, 9, 16, 25]

fig, ax = plt.subplots()
ax.plot(input_values, squares, linewidth=3)

# Set chart title and label axes.
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# Set size of tick labels.
""" 
Here's an example of other parameters you can use: ax.tick_params(axis='both', length=10, width=2, colors='r'). 
This will make the tick lines on both axes 10 points long, 2 points wide, and red in color.
"""
ax.tick_params(labelsize=14) # The size of the tick labels on both the x and y axes are set to 14 points.

plt.show()
````

````python
import matplotlib.pyplot as plt

x_values = [1, 2, 3, 4, 5]
y_values = [1, 4, 9, 16, 25]

plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s=100) #  use the s argument to set the size of the dots used to draw the graph

# Set chart title and label axes.
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# Set size of tick labels.
ax.tick_params(labelsize=14)

plt.show()
````

The `ax.ticklabel_format(style='plain')` is a way to modify the format of the labels on the tick marks (the numerical markers along the x and y axis).

Specifically, the style parameter decides the style of the numbers used for the tick labels. There are three different styles you can choose from:

`'sci'` (or 'scientific'): This uses scientific notation. For instance, 1500 would be represented as 1.5e3.
`'plain'`: This uses plain old decimal notation. For instance, 1500 would be represented as 1500.
`'': If the style argument is ''`, the power limits are used to determine the best format.

So, ax.ticklabel_format(style='plain') sets the labels on the axes to use plain decimal notation, which can be more readable in certain circumstances (like when the numbers on the axis are not too large). Without this line, Matplotlib might use scientific notation by default if the numbers on the axis are large.

````python
import matplotlib.pyplot as plt

x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s=10)

# Set chart title and label axes.
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# Set the range for each axis.
"""The axis() method requires four values: the minimum and maximum values for the x-axis and the y-axis. 
Here, we run the x-axis from 0 to 1,100 and the y-axis from 0 to 1,100,000."""
ax.axis([0, 1100, 9, 1_100_000])
ax.ticklabel_format(style='plain')

plt.show()
````

### Defining Custom Colors

To change the color of the points, pass the argument color to scatter() with the name of a color to use in quotation marks, as shown here:
```py
ax.scatter(x_values, y_values, color='red', s=10)
```
You can also define custom colors using the RGB color model. To define a color, pass the color argument a tuple with three float values (one each for red, green, and blue, in that order), using values between 0 and 1. For example, the following line creates a plot with light-green dots:
```py
ax.scatter(x_values, y_values, color=(0, 0.8, 0), s=10)
```
Values closer to 0 produce darker colors, and values closer to 1 produce lighter colors.

### Using a Colormap

A colormap is a sequence of colors in a gradient that moves from a starting to an ending color. In visualizations, colormaps are used to emphasize patterns in data. For example, you might make low values a light color and high values a darker color.

````python
import matplotlib.pyplot as plt

x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

plt.style.use('seaborn')
fig, ax = plt.subplots()
"""We pass the list of y-values to c, and then tell pyplot which colormap to use with the cmap argument. 
This code colors the points with lower y-values light blue and the points with higher y-values dark blue."""
ax.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Blues, s=10)

# Set chart title and label axes.
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# Set the range for each axis.
ax.axis([0, 1100, 9, 1_100_000])
"""The axis() method requires four values: the minimum and maximum values for the x-axis and the y-axis. 
Here, we run the x-axis from 0 to 1,100 and the y-axis from 0 to 1,100,000."""
ax.ticklabel_format(style='plain')

plt.show()
````
